# Mô hình 01 - Logistic Regression (2 nhánh)
Notebook này huấn luyện 2 nhánh:
- Nhánh A: dự báo cho người chưa mắc tiểu đường (`diabetes2 = 0`)
- Nhánh B: đánh giá nguy cơ/kiểm soát cho người đã mắc (`diabetes2 = 1`)

Target: `Glucose_class` | Weight: `std_wgt1`

## 1) Import thư viện

```python
import numpy as np
import pandas as pd
import joblib

from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import f1_score, balanced_accuracy_score, classification_report, confusion_matrix
```

## 2) Cấu hình

```python
DATA_PATH = 'Sheet1.csv'
TARGET = 'Glucose_class'
WEIGHT_COL = 'std_wgt1'
COHORT_COL = 'diabetes2'

BASE_FEATURES = [
    'Tuoi', 'gioi', 'b9', 'b11', 'b12', 'khuvuc',
    'chieucao_mean', 'cannang_mean', 'BMI', 'vongeo_mean', 'vonghong_mean', 'WHR',
    'HATT_mean', 'HATTr_mean', 'c13', 'c18', 'cvd', 'c161',
    'b10', 'c201', 'c202', 'd21', 'd24',
    'c14', 'c151', 'c152', 'c153', 'c154', 'c155', 'DT_AU_DY', 'DT_AU_TayY',
    'c162', 'c163', 'c164', 'c165',
    'c171', 'c172', 'c173', 'c174', 'c175', 'c176',
    'd22', 'd23', 'e34', 'e35', 'e36', 'e37'
],
print('Base features (Model 1 mo rong):', len(BASE_FEATURES))
```

## 3) Nạp và chuẩn bị dữ liệu
Tách các thao tác nạp dữ liệu, kiểm tra cột bắt buộc, feature engineering thành các cell riêng.

```python
# 3.1 Nạp dữ liệu thô
df = pd.read_csv(DATA_PATH)
df = df.replace({'': np.nan, ' ': np.nan, '9999': np.nan, 9999: np.nan})
print('Data shape:', df.shape)
print('Columns loaded:', df.shape[1])
```

### Giải thích bước 3.2
Kiểm tra các cột bắt buộc để tránh lỗi thiếu dữ liệu trước khi train.

```python
# 3.2 Kiểm tra cột bắt buộc
needed = BASE_FEATURES + [TARGET, WEIGHT_COL, COHORT_COL]
missing_cols = [c for c in needed if c not in df.columns]
if missing_cols:
    raise ValueError(f'Missing columns: {missing_cols}')
print('Required columns check: OK')
```

### Giải thích bước 3.3
Chuẩn hóa dữ liệu theo rule Model 1 mở rộng: map nhị phân cho biến bệnh sử/điều trị, clip theo ngưỡng, xử lý biến thai sản theo giới, và bổ sung thêm các biến tim mạch - mỡ máu để train.

```python
# 3.3 Chuẩn hóa theo đặc tả Model 1 mở rộng
numeric_cols = [
    'Tuoi', 'b10', 'b11', 'b12', 'chieucao_mean', 'cannang_mean', 'BMI',
    'vongeo_mean', 'vonghong_mean', 'WHR', 'HATT_mean', 'HATTr_mean', 'c201', 'c202',
    'c14', 'd22', 'd23', 'e34', 'e35', 'e36', 'e37'
],
for c in numeric_cols:
    if c in df.columns:
        df[c] = pd.to_numeric(df[c], errors='coerce')

# Mapping biến nhị phân theo yêu cầu
df['c13'] = df['c13'].replace({1: 1, 2: 0})
df['c18'] = df['c18'].replace({1: 1, 2: 0})
df['c161'] = df['c161'].replace({1: 0, 2: 1})
df['cvd'] = pd.to_numeric(df['cvd'], errors='coerce')

# Mapping thêm cho nhóm điều trị/tiền sử chi tiết
binary_cols = [
    'c151', 'c152', 'c153', 'c154', 'c155',
    'DT_AU_DY', 'DT_AU_TayY',
    'c162', 'c163', 'c164', 'c165',
    'c171', 'c172', 'c173', 'c174', 'c175', 'c176'
],
for c in binary_cols:
    if c in df.columns:
        df[c] = pd.to_numeric(df[c], errors='coerce').replace({1: 1, 2: 0})

# Biến thai sản theo giới
female_mask = pd.to_numeric(df['gioi'], errors='coerce') == 2
male_mask = pd.to_numeric(df['gioi'], errors='coerce') == 1
df.loc[male_mask, 'd21'] = 0
df.loc[female_mask, 'd21'] = df.loc[female_mask, 'd21'].replace({1: 1, 2: 0})
df.loc[male_mask, 'd24'] = 0
df.loc[female_mask, 'd24'] = df.loc[female_mask, 'd24'].replace({1: 1, 2: 0}).fillna(0)

# Clip theo ngưỡng được chỉ định
clip_ranges = {
    'chieucao_mean': (100, 197),
    'cannang_mean': (27, 100),
    'BMI': (14, 40),
    'vongeo_mean': (55, 128),
    'vonghong_mean': (55, 152),
    'WHR': (0.6, 1.2),
    'HATT_mean': (70, 200),
    'HATTr_mean': (50, 120),
}
for col, (lo, hi) in clip_ranges.items():
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors='coerce').clip(lower=lo, upper=hi)

# c201: 9999 đã đổi NaN ở bước nạp, tại đây đảm bảo numeric để impute median trong pipeline
df['c201'] = pd.to_numeric(df['c201'], errors='coerce')

FEATURES = BASE_FEATURES.copy()
print('Features used (Model 1 mo rong):', len(FEATURES))
```

## 5) Định nghĩa hàm huấn luyện và tuning

```python
from sklearn.model_selection import GridSearchCV

def build_model(**kwargs):
    params = {
        'max_iter': 2000,
        'class_weight': 'balanced',
        'multi_class': 'multinomial',
        'n_jobs': -1,
        'random_state': 42
    }
    params.update(kwargs)
    return LogisticRegression(**params)

def prepare_branch_data(cohort_value):
    data = df.copy()
    data[COHORT_COL] = pd.to_numeric(data[COHORT_COL], errors='coerce')
    data[TARGET] = pd.to_numeric(data[TARGET], errors='coerce')
    data = data[(data[COHORT_COL] == cohort_value) & data[TARGET].notna()].copy()
    data[TARGET] = data[TARGET].astype(int)

    cls_counts = data[TARGET].value_counts()
    keep_cls = cls_counts[cls_counts >= 10].index
    data = data[data[TARGET].isin(keep_cls)].copy()

    X = data[FEATURES]
    y = data[TARGET]
    w = pd.to_numeric(data[WEIGHT_COL], errors='coerce').fillna(1.0).clip(lower=0.0)
    return data, X, y, w

def make_pipeline(X, model):
    num_cols = [c for c in FEATURES if pd.api.types.is_numeric_dtype(X[c])]
    cat_cols = [c for c in FEATURES if c not in num_cols]

    prep = ColumnTransformer([
        ('num', Pipeline([('imp', SimpleImputer(strategy='median')), ('sc', StandardScaler())]), num_cols),
        ('cat', Pipeline([('imp', SimpleImputer(strategy='most_frequent')), ('oh', OneHotEncoder(handle_unknown='ignore'))]), cat_cols),
    ])
    return Pipeline([('prep', prep), ('model', model)])

def fit_and_evaluate(pipe, X_train, X_test, y_train, y_test, w_train, rows, classes):
    pipe.fit(X_train, y_train, model__sample_weight=w_train)
    pred_train = pipe.predict(X_train)
    pred_test = pipe.predict(X_test)

    return {
        'rows': int(rows),
        'classes': classes,
        'f1_macro_train': float(f1_score(y_train, pred_train, average='macro')),
        'f1_macro_test': float(f1_score(y_test, pred_test, average='macro')),
        'balanced_acc_train': float(balanced_accuracy_score(y_train, pred_train)),
        'balanced_acc_test': float(balanced_accuracy_score(y_test, pred_test)),
        'report_test': classification_report(y_test, pred_test),
        'cm_test': confusion_matrix(y_test, pred_test).tolist(),
    }

def train_one_branch(cohort_value, model=None):
    data, X, y, w = prepare_branch_data(cohort_value)
    X_train, X_test, y_train, y_test, w_train, _ = train_test_split(
        X, y, w, test_size=0.2, random_state=42, stratify=y
    )

    pipe = make_pipeline(X, model if model is not None else build_model())
    metrics = fit_and_evaluate(
        pipe, X_train, X_test, y_train, y_test, w_train, len(data), sorted(y.unique().tolist())
    )
    return pipe, metrics

def tune_one_branch(cohort_value, param_grid):
    data, X, y, w = prepare_branch_data(cohort_value)
    X_train, X_test, y_train, y_test, w_train, _ = train_test_split(
        X, y, w, test_size=0.2, random_state=42, stratify=y
    )

    min_class = int(y_train.value_counts().min())
    cv_folds = min(3, min_class)
    pipe = make_pipeline(X, build_model())

    if cv_folds < 2:
        fallback_metrics = fit_and_evaluate(
            pipe, X_train, X_test, y_train, y_test, w_train, len(data), sorted(y.unique().tolist())
        )
        fallback_metrics['tuned'] = False
        fallback_metrics['best_params'] = {}
        fallback_metrics['cv_folds'] = int(cv_folds)
        return pipe, fallback_metrics

    grid = GridSearchCV(
        estimator=pipe,
        param_grid=param_grid,
        scoring='f1_macro',
        cv=cv_folds,
        n_jobs=-1
    )
    grid.fit(X_train, y_train, model__sample_weight=w_train)

    best_pipe = grid.best_estimator_
    tuned_metrics = fit_and_evaluate(
        best_pipe, X_train, X_test, y_train, y_test, w_train, len(data), sorted(y.unique().tolist())
    )
    tuned_metrics['tuned'] = True
    tuned_metrics['best_params'] = grid.best_params_
    tuned_metrics['best_cv_score'] = float(grid.best_score_)
    tuned_metrics['cv_folds'] = int(cv_folds)
    return best_pipe, tuned_metrics
```

## 6) Huấn luyện baseline 2 nhánh và kiểm tra overfitting

```python
model_non_diab_base, metrics_non_diab_base = train_one_branch(0)
model_diab_base, metrics_diab_base = train_one_branch(1)

print('=== BASELINE: BRANCH non_diabetes ===')
print('F1 train/test:', metrics_non_diab_base['f1_macro_train'], metrics_non_diab_base['f1_macro_test'])
print('Balanced Acc train/test:', metrics_non_diab_base['balanced_acc_train'], metrics_non_diab_base['balanced_acc_test'])
print(metrics_non_diab_base['report_test'])

print('=== BASELINE: BRANCH diabetes ===')
print('F1 train/test:', metrics_diab_base['f1_macro_train'], metrics_diab_base['f1_macro_test'])
print('Balanced Acc train/test:', metrics_diab_base['balanced_acc_train'], metrics_diab_base['balanced_acc_test'])
print(metrics_diab_base['report_test'])
```

## 7) Hyperparameter tuning
Tối ưu tham số Logistic Regression theo từng nhánh để cải thiện khả năng tổng quát hóa.

```python
# 7.1 Khai báo lưới tham số
LR_PARAM_GRID = {
    'model__C': [0.001, 0.01, 0.1, 1, 10, 100],
    'model__solver': ['lbfgs', 'liblinear'],
    'model__max_iter': [1000, 2000, 3000]
}
LR_PARAM_GRID
```

```python
# 7.2 Tune cho 2 nhánh
model_non_diab_tuned, metrics_non_diab_tuned = tune_one_branch(0, LR_PARAM_GRID)
model_diab_tuned, metrics_diab_tuned = tune_one_branch(1, LR_PARAM_GRID)
```

```python
# 7.3 In kết quả tuning
print('=== TUNED: BRANCH non_diabetes ===')
print('Tuned:', metrics_non_diab_tuned.get('tuned'))
print('Best params:', metrics_non_diab_tuned.get('best_params'))
print('F1 train/test:', metrics_non_diab_tuned['f1_macro_train'], metrics_non_diab_tuned['f1_macro_test'])
print('Balanced Acc train/test:', metrics_non_diab_tuned['balanced_acc_train'], metrics_non_diab_tuned['balanced_acc_test'])

print('=== TUNED: BRANCH diabetes ===')
print('Tuned:', metrics_diab_tuned.get('tuned'))
print('Best params:', metrics_diab_tuned.get('best_params'))
print('F1 train/test:', metrics_diab_tuned['f1_macro_train'], metrics_diab_tuned['f1_macro_test'])
print('Balanced Acc train/test:', metrics_diab_tuned['balanced_acc_train'], metrics_diab_tuned['balanced_acc_test'])
```
