## Làm sạch dữ liệu

### Xóa các cột không có giá trị

```python
drop_cols = [
 # leakage
 'Glucose_class','Pre_diabete','predia',

 # glucose duplicate
 'glucose2','fpgva',

 # WHR duplicate
 'WHR_male','WHR_female','WHR_male2','WHR_female2','WHR_total',

 # obesity duplicate
 'obesity',

 # khu vực
 'khuvuc2',

 # tim mạch
 'cvd',

 # huyết áp
 'Nam chuan doan tang huyet ap',

 # cân nặng chuẩn
 'std_wgt1','std_wgt3',

 # text
 'Ho ten','Dia chi','Ngay phong van','Tinh/Thanhpho', 'Quan/huyen', 'Xa/phuong',

 # id
 'V1','id'
]
```

```python
# Xóa các cột
data = data.drop(columns=drop_cols)

# In danh sách cột mới
print("Danh sách cột sau khi xóa:")
print(data.columns.tolist())
```

Các cột được xóa không phải để giảm số lượng dữ liệu mà để đảm bảo mô hình học đúng và không bị sai lệch. Cụ thể:
Nhóm gây leakage: Glucose_class, predia, Pre_diabete
→ Các cột này chứa thông tin suy ra từ kết quả chẩn đoán nên gần như là đáp án. Nếu giữ lại, mô hình sẽ học “đoán trước kết quả”, dẫn đến accuracy cao nhưng không dùng được thực tế.
Nhóm trùng thông tin: WHR_male, WHR_female, WHR_male2, WHR_female2, WHR_total
→ Đều xuất phát từ cùng một chỉ số WHR, giữ nhiều sẽ làm mô hình bị lặp thông tin và dễ overfit.
Nhóm biến đổi không cần thiết: glucose2, fpgva
→ Chỉ là phiên bản transform của biến gốc, không thêm thông tin mới.
Nhóm trùng ý nghĩa: obesity, khuvuc2
→ Chỉ là dạng encode lại của BMI_class và khuvuc, nên giữ một là đủ.
Nhóm không có giá trị dự đoán: V1, id, Ho ten, Dia chi
→ Là thông tin định danh, không giúp mô hình học.
Nhóm biến suy ra: std_wgt1, std_wgt3
→ Được tính từ các biến khác nên không phải thông tin gốc.

Kết luận: các cột này bị xóa vì hoặc chứa thông tin “gần đáp án”, hoặc trùng lặp, hoặc không mang giá trị dự đoán, nhằm giúp mô hình học đúng bản chất và tránh overfit.

### Check lại dữ liệu sau khi xóa

```python
data.info()
```

```python
# ============================================================
# XỬ LÝ GIÁ TRỊ THIẾU (MISSING VALUES) - ĐÃ CẢI TIẾN
# ============================================================
# ------------------------------------------------------------
# NHÓM 1: Cột liên quan đến điều trị tăng huyết áp (~75% missing)
# Lý do: Chỉ được hỏi khi bệnh nhân CÓ tiền sử THA
# → NaN = "Không áp dụng" → Điền 0
# ------------------------------------------------------------
cols_tha = [
    'DT_THA_DY',
    'DT_THA_TY',
    'Bang an uong va luyen tap',
    'Da dieu tri tang huyet ap bao gio chua'
]
data[cols_tha] = data[cols_tha].fillna(0)
# ------------------------------------------------------------
# NHÓM 2: Cột tiền sử biến chứng & gia đình (~25% missing)
# CẢI TIẾN: Tạo cột flag đánh dấu missing thay vì chỉ điền mode
# Giúp mô hình phân biệt "không bị" vs "không có thông tin"
# ------------------------------------------------------------
cols_mode = [
    'dot_quy', 'Suy_tim', 'loet_ban_chan', 'cat_cut_chi',
    'Dau_that_nguc', 'Ong/ba noi MÐTÐ', 'Con_MÐTÐ',
    'Anh/chi/em MDTÐ', 'Bo/Me_MÐTÐ'
]
for col in cols_mode:
    # Tạo cột flag: 1 = missing, 0 = có thông tin
    data[f'{col}_missing'] = data[col].isna().astype(int)
    # Điền mode cho giá trị missing
    mode_val = data[col].mode()[0]
    data[col] = data[col].fillna(mode_val)
# ------------------------------------------------------------
# NHÓM 3: Cột mang thai (~25-34% missing)
# NaN = "Không áp dụng" (nam) → Điền 0
# ------------------------------------------------------------
data['da_mang_thai_chua'] = data['da_mang_thai_chua'].fillna(0)
data['co_bi_tieu_duong_thai_ki_khong'] = data['co_bi_tieu_duong_thai_ki_khong'].fillna(0)
# ------------------------------------------------------------
# Cột phân loại huyết áp (~30% missing)
# Dùng MEDIAN để tránh ảnh hưởng outlier
# ------------------------------------------------------------
median_val = data['Hypertension_class'].median()
data['Hypertension_class'] = data['Hypertension_class'].fillna(median_val)
# ------------------------------------------------------------
# NHÓM 4: Cột missing rất ít (<1%) → Xóa hàng
# ------------------------------------------------------------
rows_before = len(data)
data = data.dropna(subset=['GÐ_CNB_ÐTÐ', 'tien_xu_benh _tim_mach'])
rows_after = len(data)
# KIỂM TRA KẾT QUẢ
remaining_missing = data.isnull().sum().sum()
print(f"Tổng missing còn lại : {remaining_missing}")
print(f"Shape sau xử lý      : {data.shape}")
print(f"Số hàng đã xóa       : {rows_before - rows_after}")
```

```python
# ============================================================
# XỬ LÝ GIÁ TRỊ THIẾU (MISSING VALUES)
# ============================================================

# ------------------------------------------------------------
# NHÓM 1: Cột liên quan đến điều trị tăng huyết áp (~75% missing)
# Lý do: Những câu hỏi này chỉ được hỏi khi bệnh nhân CÓ tiền sử THA
# → NaN có nghĩa là "Không áp dụng" (bệnh nhân không bị THA)
# → Điền 0 = "Không / Chưa từng"
# ------------------------------------------------------------
cols_tha = [
    'DT_THA_DY',                                # Điều trị THA bằng đông y
    'DT_THA_TY',                                # Điều trị THA bằng tây y
    'Bang an uong va luyen tap',                # Điều trị bằng bảng ăn uống & luyện tập
    'Da dieu tri tang huyet ap bao gio chua'    # Đã từng điều trị THA chưa
]
data[cols_tha] = data[cols_tha].fillna(0)
# Kết quả: 4 cột từ ~75% missing → 0% missing


# ------------------------------------------------------------
# NHÓM 2: Cột tiền sử biến chứng & gia đình (~25% missing)
# Lý do: Dữ liệu bị bỏ sót trong quá trình khảo sát
# → Các cột này đều dạng nhị phân (0 = Không, 1 = Có)
# → Điền mode (giá trị xuất hiện nhiều nhất, thường là 0)
# ------------------------------------------------------------
cols_mode = [
    'dot_quy',          # Tiền sử đột quỵ
    'Suy_tim',          # Tiền sử suy tim
    'loet_ban_chan',     # Tiền sử loét bàn chân
    'cat_cut_chi',      # Tiền sử cắt/cụt chi
    'Dau_that_nguc',    # Tiền sử đau thắt ngực
    'Ong/ba noi MÐTÐ',  # Ông/bà nội mắc đái tháo đường
    'Con_MÐTÐ',         # Con mắc đái tháo đường
    'Anh/chi/em MDTÐ',  # Anh/chị/em mắc đái tháo đường
    'Bo/Me_MÐTÐ'        # Bố/mẹ mắc đái tháo đường
]
for col in cols_mode:
    mode_val = data[col].mode()[0]
    data[col] = data[col].fillna(mode_val)
# Kết quả: 9 cột từ ~25% missing → 0% missing, điền bằng mode (thường = 0)


# ------------------------------------------------------------
# NHÓM 3: Cột mang thai (~25-34% missing)
# Lý do: Bệnh nhân NAM không được hỏi câu này → NaN = "Không áp dụng"
# → Điền 0 = "Chưa từng mang thai" hoặc "Không áp dụng"
# ------------------------------------------------------------
data['da_mang_thai_chua'] = data['da_mang_thai_chua'].fillna(0)
# Kết quả: ~25% missing → 0% missing


# ------------------------------------------------------------
# Cột tiểu đường thai kỳ (~34% missing)
# Lý do: Chỉ hỏi khi bệnh nhân ĐÃ từng mang thai
# → NaN = "Không áp dụng" (chưa mang thai hoặc là nam)
# → Điền 0 = "Không bị tiểu đường thai kỳ"
# ------------------------------------------------------------
data['co_bi_tieu_duong_thai_ki_khong'] = data['co_bi_tieu_duong_thai_ki_khong'].fillna(0)
# Kết quả: ~34% missing → 0% missing


# ------------------------------------------------------------
# Cột phân loại huyết áp (~30% missing)
# Lý do: Dữ liệu đo lường bị thiếu, là cột số liên tục
# → Dùng MEDIAN thay vì mean để tránh ảnh hưởng của outlier
# → Median của Hypertension_class phản ánh mức huyết áp phổ biến nhất
# ------------------------------------------------------------
median_val = data['Hypertension_class'].median()
data['Hypertension_class'] = data['Hypertension_class'].fillna(median_val)
# Kết quả: ~30% missing → 0% missing, điền median = {median_val}


# ------------------------------------------------------------
# NHÓM 4: Cột missing rất ít (<1%) → Xóa hàng
# Lý do: Chỉ 1-21 hàng bị thiếu / 12725 tổng → không đáng kể
# → Xóa hàng an toàn hơn điền giá trị vì lượng mất rất nhỏ
# ------------------------------------------------------------
rows_before = len(data)
data = data.dropna(subset=['GÐ_CNB_ÐTÐ', 'tien_xu_benh _tim_mach'])
rows_after = len(data)
# Kết quả: Xóa {rows_before - rows_after} hàng (mất < 0.2% dữ liệu)


# ============================================================
# KIỂM TRA KẾT QUẢ SAU XỬ LÝ
# ============================================================
remaining_missing = data.isnull().sum().sum()
print(f"Tổng missing còn lại : {remaining_missing}")   # Kỳ vọng: 0
print(f"Shape sau xử lý      : {data.shape}")          # Kỳ vọng: (~12703, 51)
print(f"Số hàng đã xóa       : {rows_before - rows_after}")
```

```python
# ============================================================
# CHUYỂN ĐỔI CÁC CỘT OBJECT SANG KIỂU SỐ
# ============================================================
import re
# --- Xử lý riêng 2 cột xét nghiệm đường máu (có chứa text/ký tự đặc biệt) ---
blood_sugar_cols = ['xet_nghiem_duong_mau_luc_doi', 'xet_nghiem_duong_mau_sau_2h']
for col in blood_sugar_cols:
    # Bước 1: Chuyển sang chuỗi
    data[col] = data[col].astype(str)
    # Bước 2: Thay dấu phẩy thành dấu chấm (format số VN → Python)
    data[col] = data[col].str.replace(',', '.')
    # Bước 3: Dùng Regex trích xuất phần số (gồm số thập phân)
    data[col] = data[col].str.extract(r'(\d+\.?\d*)')[0]
    # Bước 4: Ép kiểu sang numeric
    data[col] = pd.to_numeric(data[col], errors='coerce')
# --- Ép kiểu bình thường cho các cột numeric khác ---
other_cols = [
    'chieucao_mean', 'cannang_mean', 'vongeo_mean',
    'vonghong_mean', 'HATT_mean', 'HATTr_mean',
    'BMI', 'WHR'
]
for col in other_cols:
    data[col] = pd.to_numeric(data[col], errors='coerce')
# Kiểm tra kết quả
all_converted = blood_sugar_cols + other_cols
print("Kiểu dữ liệu sau chuyển đổi:")
print(data[all_converted].dtypes)
print(f"\nSố NaN sau chuyển đổi:")
print(data[all_converted].isnull().sum())
```

```python
# ============================================================
# THỐNG KÊ MÔ TẢ CÁC CỘT LIÊN TỤC
# ============================================================
continuous_cols = [
    'chieucao_mean', 'cannang_mean', 'vongeo_mean',
    'vonghong_mean', 'HATT_mean', 'HATTr_mean',
    'BMI', 'WHR',
    'xet_nghiem_duong_mau_luc_doi', 'xet_nghiem_duong_mau_sau_2h',
    'kg_luc_nang_nhat', 'tuoi_luc_nang_nhat',
    'Thoi gian an cach bua toi'
]
print(data[continuous_cols].describe().round(2).to_string())
```

```python
# ============================================================
# XỬ LÝ NGOẠI LAI BẰNG PHƯƠNG PHÁP CAPPING (Winsorization)
# ============================================================
# Thay vì XÓA dòng chứa ngoại lai (mất dữ liệu),
# ta ĐƯA GIÁ TRỊ về ngưỡng biên bằng .clip()
# → Giữ nguyên 100% số dòng dữ liệu
continuous_cols = [
    'chieucao_mean', 'cannang_mean', 'vongeo_mean',
    'vonghong_mean', 'HATT_mean', 'HATTr_mean',
    'BMI', 'WHR',
    'xet_nghiem_duong_mau_luc_doi', 'xet_nghiem_duong_mau_sau_2h',
    'kg_luc_nang_nhat', 'tuoi_luc_nang_nhat',
    'Thoi gian an cach bua toi'
]
outlier_report = []
for col in continuous_cols:
    Q1 = data[col].quantile(0.25)
    Q3 = data[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    # Đếm ngoại lai TRƯỚC khi xử lý
    n_below = (data[col] < lower).sum()
    n_above = (data[col] > upper).sum()
    n_total = n_below + n_above
    # CAPPING: ép giá trị < lower về lower, > upper về upper
    data[col] = data[col].clip(lower=lower, upper=upper)
    outlier_report.append({
        'Cột': col,
        'Q1': round(Q1, 2),
        'Q3': round(Q3, 2),
        'IQR': round(IQR, 2),
        'Ngưỡng dưới': round(lower, 2),
        'Ngưỡng trên': round(upper, 2),
        'Dưới ngưỡng': n_below,
        'Trên ngưỡng': n_above,
        'Tổng ngoại lai': n_total
    })
# In báo cáo
report_df = pd.DataFrame(outlier_report)
print("=" * 90)
print("BÁO CÁO XỬ LÝ NGOẠI LAI - Phương pháp CAPPING (IQR × 1.5)")
print("=" * 90)
print(report_df.to_string(index=False))
print(f"\nSố dòng dữ liệu được giữ lại: {len(data)} (100%)")
print("Các giá trị ngoại lai đã được đưa về ngưỡng biên thay vì bị xóa.")
```

```python
# ============================================================
# TRỰC QUAN HÓA SAU KHI XỬ LÝ NGOẠI LAI
# ============================================================
fig, axes = plt.subplots(3, 5, figsize=(20, 12))
axes = axes.flatten()
for i, col in enumerate(continuous_cols):
    data.boxplot(column=col, ax=axes[i])
    axes[i].set_title(col, fontsize=8)
for j in range(len(continuous_cols), len(axes)):
    axes[j].set_visible(False)
plt.suptitle('Boxplot sau khi xử lý ngoại lai (Capping)', fontsize=14)
plt.tight_layout()
plt.show()
```
